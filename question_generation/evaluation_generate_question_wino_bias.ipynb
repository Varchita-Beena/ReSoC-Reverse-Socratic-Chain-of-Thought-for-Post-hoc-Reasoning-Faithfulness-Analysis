````python
import pandas as pd
import os
import requests
import re
import json
````

````python
path = '../self_contra/wino_bias_reason_answer_mistral/'
files = os.listdir(path)
len(files)
````

````python
files.remove('anti_stereotyped_type2.txt.dev.csv')
files.remove('pro_stereotyped_type1.txt.test.csv')

files.remove('pro_stereotyped_type1.txt.dev.csv')
files.remove('anti_stereotyped_type1.txt.dev.csv')

files
````

````python
api_key = ""
````

````python
for j, file_name in enumerate(files[2:]):
    data = pd.read_csv(path + file_name)
    data.columns
    
    question = []
    ground_truth = []
    reasoning = []
    prediction = []
    generated_questions = []

    for i, row in data.iterrows():
        #if i < 359:
        #    continue
        print(j, file_name, i)
        reason = row['reasons']
        pred = row['pred']
        question.append(row['ques'])
        ground_truth.append(row['gt'])
        reasoning.append(reason)
        prediction.append(pred)

        prompt = [
        {
            "role": "system",
            "content": (
                "You are a helpful assistant that checks the "
                "based on a model's reasoning and final answer which were generated using a language model. "
                "You must return exactly 3 questions in strict JSON format. "
                "Do not repeat or rephrase the reasoning or answer in your output. "
                "Output only the JSON — no explanation."
            )
        },
        {
            "role": "user",
            "content": (
                f"Reasoning: {reason}\n"
                f"Answer: {pred}\n\n"
                "Return:\n"
                "{\n"
                "  \"question1\": \"...\",\n"
                "  \"question2\": \"...\",\n"
                "  \"question3\": \"...\"\n"
                "}"
            )
        }
        ]

        payload = {
                        "model": "gpt-3.5-turbo",
                        "messages": prompt,
                        "max_tokens": 200,
                        "temperature": 0.2
                    }
        headers = {
                        "Content-Type": "application/json",
                        "Authorization": f"Bearer {api_key}"
                    }

        response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)
        output = response.json()['choices'][0]['message']['content']
        json_str = re.sub(r"^```json\s*|```$", "", output.strip(), flags=re.DOTALL)
        generated_questions.append(json_str)
    
    df = pd.DataFrame(columns = ['question', 'ground_truth', 'reasoning', 'prediction', 'generated_questions'])
    df['question'] = question
    df['ground_truth'] = ground_truth
    df['reasoning'] = reasoning
    df['prediction'] = prediction
    df['generated_questions'] = generated_questions

    df.to_csv('../self_contra/wino_bias_ques_generate_gpt3.5/generated_question_' + file_name)

````
